```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
shopping_trends = "shopping.csv"
```

```python
shopping_trends = pd.read_csv(shopping_trends)
shopping_trends
```

```python
#how many male customers vs female customers
female_count = shopping_trends[shopping_trends["Gender"] == "Female"].shape[0]
print(f"Number of females: {female_count}")
```

```python
male_count = shopping_trends[shopping_trends["Gender"] == "Male"].shape[0]
print(f"Number of males: {male_count}")
```

```python
variance = abs(male_count - female_count)

print(f"Variance between females and males: {variance}")
```

```python
labels = ['Female', 'Male']
counts = [female_count, male_count]

plt.bar(labels, counts, color=['pink', 'lightblue'])
plt.title('Variance Between Females and Males')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
#items purchased by female customers

female_item = shopping_trends[shopping_trends["Gender"] == "Female"]
f_item_count = female_item.groupby("Item Purchased")["Customer ID"].nunique()
print(f_item_count)
```

```python
#items purchased by male customers
male_item = shopping_trends[shopping_trends["Gender"] == "Male"]
m_item_count = male_item.groupby("Item Purchased")["Customer ID"].nunique()
print(m_item_count)
```

```python
#Items purchased by both male and female customers
df2 = shopping_trends.groupby(['Item Purchased', 'Gender']).size().reset_index(name='Count')
df2
```

```python
male = df2.loc[df2.Gender == "Male"]
female = df2.loc[df2.Gender == "Female"]
```

```python
#Scatter plot of items purchased by males vs Items purchased by Females
plt.scatter(male["Item Purchased"], male["Count"], c="blue", label="Male")
plt.scatter(female["Item Purchased"], female["Count"], c="pink", label="Female")
# Set labels and title
plt.xlabel('Clothing Category')
plt.ylabel('Number of Items Purchased')
plt.title('Number of Items Purchased in Each Clothing Category by Gender')

# Display legend
plt.legend(loc ="lower right")

# Rotate x-axis labels 
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.show()
```

```python
#find sum of purchases male vs female 

# Selecting only the "Gender" and "Previous Purchases" columns
purchases_df = shopping_trends[["Gender", "Previous Purchases"]]

# Displaying the new DataFrame
print(purchases_df)
```

```python
male_purchases = purchases_df[purchases_df['Gender'] == 'Male']
female_purchases = purchases_df[purchases_df['Gender'] == 'Female']
```

```python
# Display first few rows of the separated DataFrames (optional)
print("Male Purchases:")
print(male_purchases.head(10))
```

```python
print("\nFemale Purchases:")
print(female_purchases.head(10))
```

```python
plt.figure(figsize=(8, 6))

plt.boxplot([male_purchases['Previous Purchases'], female_purchases['Previous Purchases']], 
            labels=['Male', 'Female'])

plt.xlabel('Gender')
plt.ylabel('Previous Purchases')
plt.title('Box plot of Previous Purchases by Gender')

plt.show()
```

```python
male_mean = male_purchases['Previous Purchases'].mean()
male_median = male_purchases['Previous Purchases'].median()

female_mean = female_purchases['Previous Purchases'].mean()
female_median = female_purchases['Previous Purchases'].median()

# Plotting the mean and median lines
plt.figure(figsize=(8, 6))

plt.plot(['Male Mean', 'Male Median'], [male_mean, male_median], marker='o', label='Male')
plt.plot(['Female Mean', 'Female Median'], [female_mean, female_median], marker='o', label='Female')

plt.xlabel('Gender and Statistics')
plt.ylabel('Previous Purchases')
plt.title('Mean and Median of Previous Purchases by Gender')
plt.legend()

plt.show()
```

```python
# Combine male and female purchases for plotting
all_purchases = [male_purchases['Previous Purchases'], female_purchases['Previous Purchases']]

# Calculate mean and median for both genders
male_mean = male_purchases['Previous Purchases'].mean()
male_median = male_purchases['Previous Purchases'].median()

female_mean = female_purchases['Previous Purchases'].mean()
female_median = female_purchases['Previous Purchases'].median()

# Plotting all data points along with mean and median markers
plt.figure(figsize=(8, 6))

# Plot individual purchases
for i, purchases in enumerate(all_purchases):
    gender_label = 'Male' if i == 0 else 'Female'
    plt.plot([gender_label] * len(purchases), purchases, 'o', label=gender_label, alpha=0.5)

# Plot mean and median markers
plt.plot(['Male Mean', 'Male Median'], [male_mean, male_median], marker='*', label='Male Mean/Median', linestyle='', color='blue')
plt.plot(['Female Mean', 'Female Median'], [female_mean, female_median], marker='*', label='Female Mean/Median', linestyle='', color='orange')

plt.xlabel('Gender')
plt.ylabel('Previous Purchases')
plt.title('Distribution of Previous Purchases by Gender')
plt.legend()

plt.show()
```

```python
# Create subplots for male and female purchases
fig, axs = plt.subplots(1, 2, figsize=(12, 6), sharey=True)

# Calculate mean and median for both genders
male_mean = male_purchases['Previous Purchases'].mean()
male_median = male_purchases['Previous Purchases'].median()

female_mean = female_purchases['Previous Purchases'].mean()
female_median = female_purchases['Previous Purchases'].median()

# Plotting data points and mean/median markers for Male Purchases
axs[0].plot(male_purchases.index, male_purchases['Previous Purchases'], 'o', label='Male Data', alpha=0.5)
axs[0].axhline(y=male_mean, color='blue', linestyle='--', label='Male Mean')
axs[0].axhline(y=male_median, color='orange', linestyle='--', label='Male Median')

axs[0].set_xlabel('Male Purchases')
axs[0].set_ylabel('Previous Purchases')
axs[0].set_title('Distribution of Male Purchases')
axs[0].legend()

# Plotting data points and mean/median markers for Female Purchases
axs[1].plot(female_purchases.index, female_purchases['Previous Purchases'], 'o', label='Female Data', alpha=0.5)
axs[1].axhline(y=female_mean, color='blue', linestyle='--', label='Female Mean')
axs[1].axhline(y=female_median, color='orange', linestyle='--', label='Female Median')

axs[1].set_xlabel('Female Purchases')
axs[1].set_title('Distribution of Female Purchases')
axs[1].legend()

plt.tight_layout()
plt.show()
```

```python
# Selecting a small portion of data for male and female purchases
male_subset = male_purchases.head(20)
female_subset = female_purchases.head(20)

# Calculate mean and median for both genders using the subset data
male_mean = male_subset['Previous Purchases'].mean()
male_median = male_subset['Previous Purchases'].median()

female_mean = female_subset['Previous Purchases'].mean()
female_median = female_subset['Previous Purchases'].median()

# Create subplots for male and female purchases
fig, axs = plt.subplots(1, 2, figsize=(12, 6), sharey=True)

# Plotting data points and mean/median markers for Male Purchases subset
axs[0].plot(male_subset.index, male_subset['Previous Purchases'], 'o', label='Male Data', alpha=0.5)
axs[0].axhline(y=male_mean, color='blue', linestyle='--', label='Male Mean')
axs[0].axhline(y=male_median, color='orange', linestyle='--', label='Male Median')

axs[0].set_xlabel('Male Purchases')
axs[0].set_ylabel('Previous Purchases')
axs[0].set_title('Subset of Male Purchases')
axs[0].legend()

# Plotting data points and mean/median markers for Female Purchases subset
axs[1].plot(female_subset.index, female_subset['Previous Purchases'], 'o', label='Female Data', alpha=0.5)
axs[1].axhline(y=female_mean, color='blue', linestyle='--', label='Female Mean')
axs[1].axhline(y=female_median, color='orange', linestyle='--', label='Female Median')

axs[1].set_xlabel('Female Purchases')
axs[1].set_title('Subset of Female Purchases')
axs[1].legend()

plt.tight_layout()
plt.show()
```
